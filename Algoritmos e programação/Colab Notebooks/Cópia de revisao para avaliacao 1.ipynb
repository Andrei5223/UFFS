**1** - O **Algoritmo de Euclides** está presente no quinto livro de Euclides, escrito por volta de 300 anos antes de Cristo. Sua finalidade é calcular o Máximo Divisor Comum (M.D.C.), sendo um algoritmo que desempenha importante papel na matemática.

![](https://drive.google.com/uc?export=view&id=1Erql3qFWzBEaGxT-7VAoTfVae7HrB32n)

Segue-se o algoritmo:
1.  Leia duas variáveis a e b
2.  r = resto da divisão entre a e b
3.  a = b
4.  b = r
5.  Se o valor de **r** é zero, então **a** é o M.D.C. procurado e o programa termina; caso contrário, volte ao passo 2.

````python
a = int(input())
b = int(input())
while True:
  r = a % b
  if r == 0:
    break
  else:
    a = b
    b = r
print(b)
````

**2** - Faça um programa que leia 5 valores inteiros positivos. A cada valor lido, o programa deve mostrar na tela a sequência de valores pares existentes entre 0 e o respectivo valor.

Ex: leu o valor 13

Exibe: 0 2 4 6 8 10 12

````python
for i in range(0, 5):
  a = int(input('Insira um valor: '))
  for s in range(0, (a + 1), 2):
    print(s, end = ' ')
  print()
````

**3** - Faça a seguinte figura usando While encadeado e com apenas um caractere * no print:
```
*
* *
* * *
* * * *
```

````python
a = 0
b = 0
while a < 4:
  a += 1
  b = 0
  while b < a:
    print('*', end = ' ')
    b += 1
  print()
````

**4** - Faça um programa que leia as idades de um conjunto de 10 pessoas, calcule e exiba a idade média do grupo.

Após a leitura cada idade, esta deve ser validada: enquanto o valor estiver fora do intervalo 0 a 120, o programa deve mostrar uma mensagem "idade inválida" e solicitar novamente a digitação. Somente após informar um valor dentro do intervalo é que o programa prossegue com a realização dos cálculos.

````python
while True:
  soma = 0
  for i in range(1, 11):
    idade = int(input(f'Insira a idade da {i}° pessoa: '))
    soma += idade
  media = soma / 10
  if media < 0 or media >= 120:
    print('Idade inválida.\nInsira novas idades.')
    continue
  else:
    print(f'A idade média é: {media:.2f}')
    break
````
